# Lab 3 - Video Game Reviews

```python
import pandas as pd
```

## Import Data

```python
df = pd.read_csv('imdb-videogames.csv')
df.drop_duplicates(inplace=True)
df.rating.replace(' ', pd.NA, inplace=True)
df.rating.replace('*9.0', '9', inplace=True)
df.rating = df.rating.astype(float)
df.votes.replace(' ', pd.NA, inplace=True)
df.votes = df.votes.str.replace(',', '')
df.votes = df.votes.astype(float)
df.drop('Unnamed: 0', axis = 1, inplace=True)
df.rename(columns={'plot summary':'plot_summary'}, inplace=True)
```

## Questions

### 1. What is the average rating for Action games?

```python
df[df.Action == True].rating.mean()
```

### 2. What is the average rating of games after 2010? What is the average rating of games before (and including) 2010?

```python
df[df.year >= 2010].rating.mean()
```

### 3. What games include "Fallout" in their title?

```python
df[df.name.str.contains("Fallout")]
```

### 4. What is the most common certificate for Sci-Fi games? 

```python
df[df["Sci-Fi"]].certificate.mode().loc[0]
```

### 5. What percent of Thiller games have M or AO certificate?

```python
(df.Thriller[(df.certificate == 'M')|(df.certificate == 'AO')].sum() / df.Thriller.sum())* 100
```

### 6. What is the mean and median rating for Family games? for Action games?

```python
print(df[df.Family].rating.agg({'family_mean':'mean','family_median':'median'}))
df[df.Action].rating.agg({'action_mean':'mean','action_median':'median'})
```

### 7. Which game from 2018 got the most votes? 

```python
df[df.year==2018][['name', 'votes']].max()
```

### 8. What is the average rating for each genre (Action, Adventure, Comedy, Crime, Family, Fantasy, Mystery, Sci-Fi, and Thriller)?

```python
print('Action:',df[df.Action].rating.mean())
print('Adventure:',df[df.Adventure].rating.mean())
print('Comedy:',df[df.Comedy].rating.mean())
print('Crime:',df[df.Crime].rating.mean())
print('Family:',df[df.Family].rating.mean())
print('Fantasy:',df[df.Fantasy].rating.mean())
print('Mystery:',df[df.Mystery].rating.mean())
print('Sci-Fi:',df[df['Sci-Fi']].rating.mean())
print('Thriller:',df[df.Thriller].rating.mean())
```

### 9. Which year's games received the most votes?

```python
df.groupby('year')['votes'].sum().sort_values(ascending= False)
```

### 10. What's the average number of votes for games that are not Action games?

```python
df[~df.Action][['votes']].mean()
```

### 11. What game are platformers?

```python
df[df.plot_summary.str.contains("platformer")]
```

12. What is the most common certificate of games with ratings greater than 8.0 and votes greater than 1000?


```python
df[(df.rating > 8) & (df.votes > 1000)].certificate.mode().loc[0]
```

13. For games with ratings greater than the 75th percentile for ratings and a number of votes greater than the 75th percentile for votes, what is the most common genre?


```python
good_games = ['Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy', 'Mystery', 'Sci-Fi', 'Thriller']
for game in good_games:
    print(game +':' , df[(df.rating > df.rating.quantile(0.75)) & (df.votes > df.votes.quantile(.75)) & (df[game])][game].sum())
```

14. What are the average ratings by year?


```python
df.groupby('year')['rating'].mean()
```

15. For each cerficate, what was the average rating and average number of votes?


```python
df.groupby('certificate')['rating'].mean()
```
